# Student Marks Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Loading Sample Data

```python
student_data = pd.read_csv('students.csv')
```

```python
student_data
```

### Visulaization of data

```python
marks = list(range(55,105,5))
len(marks)
```

```python
test = {"Test1":"red","Test2":"blue","Test3":"orange","Test4":"purple","Test5":"black",
       "Test6":"violet", "Test7":"pink", "Test8":"green","Test9":"yellow","Test10":"Brown"}
plt.scatter(student_data.Tests, student_data.Maths, c=student_data.Tests.map(test))
plt.xlabel("No of Tests")
plt.ylabel("Maths Marks")
plt.show()
```

### Creating the histogram for subjects

```python
bins = list(range(10,110,10))
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(7,7))
ax1.hist(student_data.Maths, bins)
ax1.set_xlabel("Maths")
ax2.hist(student_data.English, bins)
ax2.set_xlabel("English")
ax3.hist(student_data.Aptitude, bins)
ax3.set_xlabel("Aptitude")
ax4.hist(student_data.Science, bins)
ax4.set_xlabel("Science");
```

```python
plt.hist(student_data.Computer, bins)
plt.xlabel("Computer")
plt.show()
```

```python
student_data.plot()
```

```python
plt.figure(figsize=(7,7))
plt.scatter(student_data.Tests, student_data.Maths, c='Red', label="Maths")
plt.scatter(student_data.Tests, student_data.Science, c='Blue', label="Science")
plt.scatter(student_data.Tests, student_data.English, c='green', label="English")
plt.scatter(student_data.Tests, student_data.Aptitude, c='violet', label="Aptitude")
plt.scatter(student_data.Tests, student_data.Computer, c='yellow', label="Computer")
plt.legend();
```
